## 实验二：34个中国省会城市、直辖市、特别行政区

### 定义画图函数 

```python
# 线条的颜色(默认：飞机的架数不超过颜色的种数)
line_color = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'darkblue', 'darkgreen', 
              'cadetblue', 'pink', 'lightblue', 'lightgreen', 'gray', 'black']

# 把汉字从utf-8编码为ascii xml字符的形式
def utf2asc(s):
    return str(str(s).encode('ascii', 'xmlcharrefreplace'))[2:-1]

def get_all_path_city_loc(obj):
    if isinstance(obj, GA):
        best_path = obj.best_path  # 取别名：最优个体的路径组成
    elif isinstance(obj, SA):
        best_path , best_dis = obj.get_china_city_distance(obj.best_solution)
        
    # 取别名：所有城市的地理位置构成 如cities[1] = ['北京', 116.407526, 39.90403]。城市下标从1开始
    cities = obj.china_cities  
    all_path_loc = []
    for i in range(len(best_path)):
        one_path = []
        if len(best_path[i]) == 0:  # 这架飞机不出发，不做另外处理
            pass
        else:
            # 比如：one_path = [ [22.73264868398435, 120.28450012207031],[22.72837380478485, 120.28450012207031]]
            # 添加起点
            one_path.append([cities[obj.start_index][2], cities[obj.start_index][1]] )
            for j in range(len(best_path[i])):
                one_path.append([cities[best_path[i][j]][2], cities[best_path[i][j]][1]] )
                if j == len(best_path[i]) - 1:
                    # 添加终点
                    one_path.append([cities[obj.start_index][2], cities[obj.start_index][1]] )
            all_path_loc.append(one_path)
                    
    return all_path_loc
    

def plot_all_best_path(obj):
    """
    画航线地图
    Args:
            obj:obj对象
    Returns:
            m:地图        
    """
    # 注意：location的格式为[纬度,经度]
    start_city_loc = [obj.china_cities[obj.start_index][2], obj.china_cities[obj.start_index][1]]
    # 所有路线组成
    all_path_loc = get_all_path_city_loc(obj)
    # 设置地图
    m = folium.Map(location=start_city_loc,
                   zoom_start=5,
                   control_scale=True)
     # 设置起点的Marker
    icon_kw = dict(prefix='fa', color='orange', icon_color='darkred', icon='fa-star')
    icon = folium.Icon(**icon_kw)
    # popup：鼠标点击显示信息；tooltip：鼠标放上去显示信息
    
    folium.Marker(
        location=start_city_loc,
        # 点击起点，显示城市名
        popup="<b>{}</b>".format(utf2asc(obj.china_cities[obj.start_index][0])),
        icon=icon
    ).add_to(m);
    
    l_color = line_color[:]
    for i in range(len(all_path_loc)):
        # 打乱颜色
        random.shuffle(l_color)
        r_str = '第{}架运输机航线'.format(i+1)
        ls = folium.PolyLine(locations=all_path_loc[i],
                             weight=2,
                             popup="<b>{}</b>".format(utf2asc(r_str)),  # 点击航线，显示航线信息
                             tooltip="<b>{}</b>".format(utf2asc(r_str)),  # 鼠标悬停航线，显示航线信息
                             color=l_color[i])
        ls.add_to(m)
    return m
```

### 定义GA对象

```python
# 定义相关参数
salesman_num = 4
start_index = 17  # 起始城市下标
city_num = 34
bal_weight = 4000

china_ga_obj = GA(salesman_num=salesman_num,
            city_num=city_num,
            start_index=start_index,
            filename="./data/provincial_capital.txt",
            max_gen=1000,
            distance_weight=1,
            balance_weight=bal_weight)

```

### 运输机最佳航线迭代过程

```python
# 执行遗传算法
china_ga_obj.china_city_ga_process()
```

### 最佳航线

```python
# 打印全局最优个体的最佳路径
china_ga_obj.print_china_city_best_routine()
```

### 最优个体总路程变化情况 

```python
# 全局最优个体和每代最优个体的总路程变化情况
china_ga_obj.plot_dis_sum_diff()
```

### 最优个体适应度变化情况 

```python
# 全局最优个体和每代最优个体适应度变化情况
china_ga_obj.plot_chrom_fit_diff()
```

### 地图显示最佳航线

```python
# 路径可视化
ga_city_path = plot_all_best_path(china_ga_obj)
```

```python
# 显示地图
ga_city_path
```

```python
# 保存地图为html文件
ga_city_path.save('./airline/ga_best_all_air_line.html')
```
